#Time Series Analysis of Beef 


```
import numpy as np
import pandas as pd
url='https://raw.githubusercontent.com/rohittttt28/Time_seriesbasics/master/meat_prices.csv'
```

```
data = pd.read_csv(url)
data.describe()
print("\n\n\nDataset has {} rows and {} columns".format(data.shape[0], data.shape[1]))
print("\n\n\nDatatype: \n")
print(data.dtypes)
print("\n\n\nThe number of null values for each column are: \n")
print(data.isnull().sum())
print("\n\n\nData summary: \n")
print(data.describe())
```

```
data.Month  = pd.to_datetime(data.Month.str.upper(), format='%b-%y', yearfirst=False)
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data.dropna(inplace=True)
```

```
data.shape[0]
data.head()
data = data.replace('%', '', regex=True)
data["Beef price % Change"] = data["Beef price % Change"].astype("float")
```

```
plt.figure(num=None, figsize=(15, 11), dpi=80)
plt.ylabel('Price  ')
plt.xlabel('Months')
#plt.title('  Chicken Price ')
plt.plot(data["Month"].tolist(),data["Beef Price"].tolist(), c='Blue', alpha = 1, label = 'Beef Price (1990-2020)')
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    

plt.grid(False)
plt.grid(linestyle= 'dashed', linewidth=1.5, axis = 'y')
sns.set_style("whitegrid")
plt.legend()
plt.show()
plt.figure(num=None, figsize=(15, 11), dpi=80)
plt.ylabel('Price  ')
plt.xlabel('Months')
#plt.title('  Chicken Price ')
plt.plot(data["Month"].tolist(),data["Beef price % Change"].tolist(), c='black', alpha = 1, label = 'Beef Price % change(1990-2020)')
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    

plt.grid(False)
plt.grid(linestyle= 'dashed', linewidth=1, axis = 'y')
sns.set_style("whitegrid")
plt.legend()
plt.show()
```
